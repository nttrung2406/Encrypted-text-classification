```python
import pandas as pd
import os
from scipy.stats import mode
```

```python
root = os.getcwd()
result_path = os.path.join(root, "all_model_predictions.csv")
result = pd.read_csv(result_path)
result
```

```python
model_columns = ["DistilBERT_predictions", "RoBERTa_predictions", "XLNet_predictions", "ELECTRA_predictions", "DeBERTa_predictions"]
result["labels"] = result[model_columns].mode(axis=1)[0]
```

```python
df = result.drop(columns=["DistilBERT_predictions", "RoBERTa_predictions", "XLNet_predictions", "ELECTRA_predictions", "DeBERTa_predictions"])
df["labels"] = df["labels"].astype(int)
df
```

```python
with open('predictions.txt', 'w') as f:
    for index, row in df.iterrows():
        tokens = row['texts'] 
        predicted_label = row['labels']  
        formatted_output = f"'{tokens}' {predicted_label}\n"
        
        f.write(formatted_output)

print("Predictions have been saved to predictions.txt")
```
